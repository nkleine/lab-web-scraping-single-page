```python
# Build Table with all Artist Names and Titles:
```

```python
Artist = []
Title = []
#Genre = []
```

```python
num_iter = len(soup.select("p.title-artist cite.title"))

TitleList = soup.select("p.title-artist cite.title")
ArtistList = soup.select("p.title-artist em.artist")
#GenreList = soup.select("div.chart-content.col-xs-12.col-sm-8 > ul a")
```

```python
# iterate through the result set and retrive all the data
for i in range(num_iter):
    Title.append(TitleList[i].get_text())
    Artist.append(ArtistList[i].get_text())
    #Genre.append(GenreList[i].get_text())

print(Title)
print(Artist)
#print(Genre)
```

```python
# each list becomes a column
Top100 = pd.DataFrame({'Artist':Artist, "Title":Title})
```
